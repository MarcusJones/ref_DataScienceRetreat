# Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
train.target.value_counts().plot(kind='bar')
sns.despine()

#the class distribution is balanced. equal amount of samples of classes.
```

```python
train.hist(figsize=(28,28))
sns.despine()

#histogram of the feature values show that some features are left and some others have right skewed distribution.
#this may imply that features have a correlation with each other. 
```

```python
feature_cols = [f for f in train.columns if "feature" in f]
```

```python
sns.pairplot(train, vars=feature_cols[0:5], kind="reg",size=5)
sns.despine()

#when we plot pair plot we see that some features are linearly corraleted with each other. 
#for example f4 and f2 have a negative linear relation 
```

```python
sns.pairplot(train, vars=feature_cols[0:5], kind="reg",size=5,hue='target')
```

# visualise label distribution for each Numerai feature

```python
target_0 = train.loc[train['target'] == 0]
target_1 = train.loc[train['target'] == 1]

for feature in feature_cols:
    plt.figure(figsize=(8,8))
    ax=sns.distplot(target_0.loc[:,feature], hist=False)
    ax=sns.distplot(target_1.loc[:,feature], hist=False)    
    plt.show()
```

```python
max_features = 10000
max_len = 500
batch_size = 32

(x_train1, y_train), (x_test1,y_test) = ks.datasets.imdb.load_data(num_words = max_features)
```

```python
print(x_train1.shape)
print(x_test1.shape)
```

```python
#import keras.preprocessing as preproc
#print(preproc)

x_train = ks.preprocessing.sequence.pad_sequences(x_train1,maxlen = max_len)
x_test = ks.preprocessing.sequence.pad_sequences(x_test1,maxlen = max_len)
```

```python
model = ks.models.Sequential()
model.add(ks.layers.Embedding(max_features,32)) # Each word is a vector of len 32
model.add(ks.layers.SimpleRNN(32))
model.add(ks.layers.Dense(1, activation="sigmoid"))
model.summary()
```

```python
model.compile(  
    optimizer = ks.optimizers.RMSprop(lr=0.0001),
    loss= ks.losses.binary_crossentropy,
    metrics= ["accuracy"],
)
```

```python
checkpointer = ks.callbacks.ModelCheckpoint(filepath=r'C:\Temp DEL checkpoint\weights.hdf5', verbose=1, save_best_only=True)
```

```python
history = model.fit(
    x_train, y_train,
    epochs = 10,
    batch_size = 128,
    validation_split = 0.2,
    callbacks=[checkpointer],
)
```

```python
def plot_hist(history):
   fig = plt.figure(figsize=(5,4))
   f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5), sharey=False)
   
   ax1.plot(history.epoch,  history.history['loss'], label="Train")
   ax1.plot(history.epoch,  history.history['val_loss'], label="CV")
   ax1.set_title("Loss function development - Training set vs CV set")
   ax1.legend(loc='upper right')
   ax1.set_xlabel('Epochs')
   ax1.set_ylabel('Values')
   
   ax2.plot(history.epoch,  history.history['acc'],label="Train")
   ax2.plot(history.epoch,  history.history['val_acc'],label="CV")
   ax2.set_title("Accuracy development - Training set vs CV set")
   ax2.legend(loc='upper right')
   ax2.set_xlabel('Epochs')
   ax2.set_ylabel('Values')
   
   #plt.suptitle(model_title, fontsize=16)
   
   plt.show()


plot_hist(history)
```
